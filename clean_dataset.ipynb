```python
import datetime

date_needed = datetime.datetime(2022, 12,3,0,0)

full_date = "%d %B %Y"
def suffix(day):
    suffix = ""
    if 4 <= day <= 20 or 24 <= day <= 30:
        suffix = "th"
    else:
        suffix = ["st", "nd", "rd"][day % 10 - 1]
    return suffix

my_date = date_needed.strftime("%d" + suffix(date_needed.day) + " %B %Y")
print(my_date)

print(f"Full date: {date_needed.strftime(full_date)}.")
```
